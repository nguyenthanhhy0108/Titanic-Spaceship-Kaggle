```python
import numpy as np
import matplotlib as plt
import pandas as pd
```

```python
def get_first_name(sentence):
  name = sentence.split(' ')
  return name[len(name) - 1]
```

```python
def cabin_to_int(cabin):
    s1 = cabin[0]
    a = ord(s1)
    s3 = cabin[len(cabin) - 1]
    c = ord(s3)
    b = int(cabin[2:len(cabin) - 2])
    return a, b, c
```

```python
def get_last_name(sentence):
  name = sentence.split(' ')
  return name[0]
```

```python
def count_num(number, arr):
  count = 0
  for i in range(len(arr)):
    if arr[i] == number:
      count += 1
  return count
```

```python
arr = [1, 3, 4, 6]
for j in range(len(arr)):
  k = arr[j]
  for row in range(len(df)):
    if pd.isnull(df.iloc[row, k]) == True and pd.isnull(df.iloc[row, 12]) != True:
      this_fname = get_first_name(df.iloc[row, 12])
      for i in range(len(df)):
        if(pd.isnull(df.iloc[i, k]) != True and pd.isnull(df.iloc[i, 12]) != True):
          if get_first_name(df.iloc[i, 12]) == this_fname:
            df.iloc[row, k] = df.iloc[i, k]
            break
  for row in range(len(df)):
    if pd.isnull(df.iloc[row, k]) == True and pd.isnull(df.iloc[row, 12]) != True:
      this_fname = get_last_name(df.iloc[row, 12])
      for i in range(len(df)):
        if(pd.isnull(df.iloc[i, k]) != True and pd.isnull(df.iloc[i, 12]) != True):
          if get_last_name(df.iloc[i, 12]) == this_fname:
            df.iloc[row, k] = df.iloc[i, k]
            break
```

```python
for row in range(len(df)):
  sum = []
  sum.append(df.iloc[row, 7])
  sum.append(df.iloc[row, 8])
  sum.append(df.iloc[row, 9])
  sum.append(df.iloc[row, 10])
  sum.append(df.iloc[row, 11])
  if count_num(0, sum) > 2:
    df.iloc[row, 7] = df.iloc[row, 8] = df.iloc[row, 9] = df.iloc[row, 10] = df.iloc[row, 11] = 0
```
